```python
documents=[]
for i in range(df.shape[0]):
    line=df.iloc[i]
    txt=word_tokenize(line[1])
    r=line[0]
    tup=(txt,r)
    documents.append(tup)
    
documents[0:5]
```

```python
stops=stopwords.words('english')
punctuations=list(string.punctuation)
stops.extend(punctuations)
print(len(stops))
stops
```

```python
lemmatizer=WordNetLemmatizer()
lemmatizer.lemmatize(documents[0][0][5])
```

```python
def simple_pos(tag):
    if tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
        
```

```python
simple_pos('JJ')
```

```python
lemmatizer.lemmatize('better',pos=simple_pos(pos_tag(['better'])[0][1]))
```

```python
def clean_review(words):
    op=[]
    for w in words:
        if w not in stops:
            p=pos_tag([w])
            cw=lemmatizer.lemmatize(w,pos=simple_pos(p[0][1]))
            op.append(cw)
    return op
```

```python
documents=[(clean_review(doc),tar) for doc,tar in documents]
documents[0:5]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
import random
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
```

```python
random.shuffle(documents)
random.shuffle(documents)
random.shuffle(documents)
random.shuffle(documents)
documents[0:5]
```

```python
vec_documents=[' '.join(doc) for doc,tar in documents]
vec_documents[0:5]
```

```python
targets=[tar for doc,tar in documents]
targets[0:5]
```

```python
vec=TfidfVectorizer(max_features=5000,ngram_range=(1,3))
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(vec_documents,targets,random_state=7)
```

```python
xtrain_transform=vec.fit_transform(xtrain)
```

```python
xtest_transform=vec.transform(xtest)
```

```python
xtrain_transform.todense()
```

```python
clf=MultinomialNB()
```

```python
clf.fit(xtrain_transform,ytrain)
```

```python
predictions=clf.predict(xtest_transform)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(classification_report(ytest,predictions))
print(confusion_matrix(ytest,predictions))
```
